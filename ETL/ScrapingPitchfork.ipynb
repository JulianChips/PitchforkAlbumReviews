```python
# Dependencies
import os
import requests
from bs4 import BeautifulSoup as bs
import pymongo
import json
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
pf_db = client.pitchfork
```

```python
reviews = pf_db.album_reviews
```

```python
#Plans:
#Loop Through each page of most recent Pitchfork reviews
#Find links to every review
#scrape album name, artist name, review score, review body
#store these reviews in a database or file to do sentiment analysis
```

```python
# URL of page to be scraped
url = 'https://pitchfork.com/reviews/albums/?page=1'
```

```python
# Retrieve page with the requests module
response = requests.get(url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = bs(response.text, 'html.parser')
```

```python
# results are returned as an iterable list
results = soup.find_all('a', class_="review__link")
```

```python
print(results[0]["href"])
```

```python
review_url = 'https://pitchfork.com' + results[0]["href"]
```

```python
review_response = requests.get(review_url)
```

```python
review_soup = bs(review_response.text, 'html.parser')
```

```python
review_results = review_soup.find_all('p')
```

```python
review_text = ""
for item in review_results[0:len(review_results)-2]:
    review_text = review_text+item.text
print(review_text)
```

```python
artist_name = review_soup.find("ul",class_="artist-links").find("a").text
print(artist_name)
```

```python
album_name = review_soup.find("h1",class_="single-album-tombstone__review-title").text
print(album_name)
```

```python
review_score = float(review_soup.find("span",class_="score").text)
print(review_score)
```

```python
genres = review_soup.find_all("li",class_="genre-list__item")
```

```python
genre_list = []
for genre in genres:
    genre_list.append(genre.text)
print(genre_list)
```

```python
review_insert = {
    "artist":artist_name,
    "album":album_name,
    "review":review_text,
    "score":review_score,
    "genres":genre_list
}
```

```python
reviews.insert(review_insert)
```

```python
for item in reviews.find():
    print(item)
```

```python
for page in range(1,20):
    url = 'https://pitchfork.com/reviews/albums/?page='+str(page)
    response = requests.get(url)
    soup = bs(response.text, 'html.parser')
    results = soup.find_all('a', class_="review__link")
    for result in results:
        print(result["href"])
        url = 'https://pitchfork.com' + result["href"]
        response = requests.get(url)
        soup = bs(response.text, 'html.parser')
        review_results = soup.find_all('p')
        review_text = ""
        for item in review_results[0:len(review_results)-2]:
            review_text = review_text+" "+item.text
        try:
            artist_name = soup.find("ul",class_="artist-links").find("a").text
        except:
            artist_name = "Various Artists"
        album_name = soup.find("h1",class_="single-album-tombstone__review-title").text
        review_score = float(soup.find("span",class_="score").text)
        genres = soup.find_all("li",class_="genre-list__item")
        genre_list = []
        for genre in genres:
            genre_list.append(genre.text)
        review_insert = {
            "artist":artist_name,
            "album":album_name,
            "review":review_text,
            "score":review_score,
            "genres":genre_list
        }
        reviews.insert(review_insert)
```
